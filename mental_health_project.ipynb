```python
%%writefile mental_health_project/utils/bot_logic.py
import streamlit as st

def bot_reply(msg):
    msg = msg.lower()

    if "current_mood" not in st.session_state:
        st.session_state.current_mood = ""

    # Crisis
    if any(w in msg for w in ["suicide", "kill myself", "want to die", "end my life"]):
        return (
            "ðŸ›‘ I'm really sorry you're feeling this way.\n\n"
            "ðŸ“ž **KIRAN (24/7): 1800-599-0019**\n"
            "ðŸ“ž **AASRA: +91-9820466726**\n\n"
            "You are not alone."
        )

    if "breakup" in msg:
        st.session_state.current_mood = "breakup"
        return "ðŸ’” Breakups hurt deeply. What part hurts the most?"

    if "sad" in msg:
        st.session_state.current_mood = "sad"
        return "ðŸ˜” I'm really sorry you're feeling sad."

    if "happy" in msg:
        return "ðŸ˜Š I'm glad you're feeling happy!"

    return "ðŸ¤– I'm here with you. You can share more."
```

```python
%%writefile mental_health_project/pages/1_Chatbot.py
import streamlit as st
from utils.bot_logic import bot_reply

st.title("ðŸ’¬ Mental Health Chatbot")

if "chat" not in st.session_state:
    st.session_state.chat = []

user_input = st.text_input("Share your feelings:")

if user_input:
    st.session_state.chat.append(("You", user_input))
    st.session_state.chat.append(("Bot", bot_reply(user_input)))

for sender, msg in st.session_state.chat:
    st.write(f"**{sender}:** {msg}")
```

```python
%%writefile mental_health_project/pages/2_Mood_Tracker.py
import streamlit as st
import matplotlib.pyplot as plt
import datetime

st.title("ðŸ“Š Mood Tracker")

if "mood" not in st.session_state:
    st.session_state.mood = []

if st.button("Add Sample Mood"):
    st.session_state.mood.append((datetime.datetime.now(), 5))

if len(st.session_state.mood) > 1:
    times = [m[0] for m in st.session_state.mood]
    scores = [m[1] for m in st.session_state.mood]

    fig, ax = plt.subplots()
    ax.plot(times, scores, marker="o")
    ax.set_ylim(0, 10)
    st.pyplot(fig)
```
